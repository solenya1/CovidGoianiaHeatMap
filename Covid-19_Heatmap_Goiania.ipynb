<h2> COVID-19 GYN HEATMAP </H2>
making a covid-19 heatmap for Goîania , showing the neighborhoods whit the highest cofirmations cases.

```python
import pandas as pd
import numpy as np
import folium
from folium import plugins
```

```python
# lat long goiania : 16.6869° S, 49.2648° W
#goiania = folium.Map(location=[-16.6799, -49.255], zoom_start=12)
#import the dataset
df = pd.read_csv('distribuicao_covid_gyn.csv')
df
```

```python
#Separeted only the coordinates and cases
coordinates = df[['latitude' , 'longitude', 'casos']]
test_cord = coordinates[0:85]
test_cord
```

<html>
    <h2> Heat Map </h2>
    </html>

```python
#creating the map using <folium>
baseMap = folium.Map(
                width="100%",
                height="100%",
                location=[-16.6799, -49.255],
                zoom_start=12
)
```

```python
baseMap
```

```python
#Drawing the heatmap on the map selected
baseMap = baseMap.add_child(plugins.HeatMap(test_cord))
```

```python
baseMap
```

```python
df['bairros'][4] = 'Parque Amazonia' # just make some corretions in the name 
```

```python
df = df.dropna() # removing the NaN on the Dataset
df
```

```python
#Creating a circle with the information about the cases in the neighborhood
for i in range(0, len(df)):
    folium.Circle(
        location = [df.iloc[i]['latitude'], df.iloc[i]['longitude']],
        color   = '#00FF69',
        fill    = '#00A1B3',
        tooltip = '<li><bold> Bairro:' + str(df.iloc[i]['bairros']) +
                  '<li><bold> Casos: ' + str(df.iloc[i]['casos']),
        radius  = (df.iloc[i]['casos']**1.1)
    ).add_to(baseMap)
```

```python
baseMap # and here it is !
```

```python
baseMap.save('index.html')
```
